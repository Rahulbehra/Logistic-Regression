```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y=make_classification(n_features=2,n_informative=2,n_samples=200,n_redundant=0,n_classes=2,hypercube=2,n_clusters_per_class=1,random_state=1)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
p_model=PolynomialFeatures(degree=3,include_bias=False)
p_data=p_model.fit_transform(x)
```

```python
model=LogisticRegression()
model.fit(p_data,y)
```
