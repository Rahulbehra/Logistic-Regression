```python
from sklearn.datasets import make_classification
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_clusters_per_class=1,n_redundant=0,n_classes=2,hypercube=False,class_sep=20,random_state=41)
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
def sigmod(x):
    return 1/(1+np.exp(-x))
```

```python
def gd(x,y):
    x=np.insert(x,0,1,axis=1)
    weight=np.ones(x.shape[1])
    lr=0.5
    for i in range(5000):
        y_hat=sigmod(np.dot(x,weight))
        weight+=lr*(np.dot((y-y_hat),x)/x.shape[0])
    return weight[1:],weight[0]
```

```python
coef_,intercept=gd(x,y)
```

```python
coef_
```

```python
intercept
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
```

```python
# plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.plot(x_input,y_input,c='red',linewidth=3)
plt.ylim(-3,2)
```
