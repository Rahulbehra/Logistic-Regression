```python
from sklearn.datasets import make_classification
import numpy as np
x,y=make_classification(n_samples=100,n_features=2,n_informative=1,n_clusters_per_class=1,n_redundant=0,n_classes=2,hypercube=False,class_sep=30,random_state=41)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter')
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    weight=np.ones(x.shape[1])
    lr=0.1
    for i in range(100):
        j=np.random.randint(1,100)
        j_hat=sigmoid(np.dot(x[j],weight))
        weight+=lr*(y[j]-j_hat)*x[j]
    return weight[0],weight[1:]
```

```python
intercept_,coef_=perceptron(x,y)
```

```python
m=-(coef_[0]/coef_[1])
b=-(intercept_/coef_[1])
```

```python
x_input=np.linspace(-3,3,100)
y_input=m*x_input+b
y_input
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.ylim(-5,5)
```
